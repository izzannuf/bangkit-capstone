```python
import pandas as pd
```

```python
df = pd.read_csv("dataset.csv")
df
```

```python
# Drop the alamat and kota_administrasi columns
df.drop(["alamat", "kota_administrasi"], axis=1, inplace=True)
df
```

```python
# Add an ID Column that is an incrementing integer from 1
df.insert(0, 'id', range(1, 1 + len(df)))
df
```

```python
# Add a column called url_image and set all values to NaN
df.insert(5, 'url_image', float('nan'))
df
```

```python
# Change id column name to id_toko
df.rename(columns={'id': 'id_toko'}, inplace=True)
df
```

```python
# Reorder the columns to be id_toko, nama_toko, kategori, latitude, longitude, deskripsi, url_image, rerata_rating, jumlah_rating
df = df[['id_toko', 'nama_toko', 'kategori', 'latitude', 'longitude', 'deskripsi', 'url_image', 'rerata_rating', 'jumlah_rating']]
df
```

```python
# Save the dataframe to a new csv file called dataset_final.csv
df.to_csv("dataset_final.csv", index=False)
```

```python
import pandas as pd
```

```python
df = pd.read_csv("Lokasi-Binaan-Pedagang-Kaki-Lima-Di-DKI-Jakarta-.csv")
df
```

```python
# Remove status_jakcard, no_kios columns
df = df.drop(['status_jakcard', 'no_kios'], axis=1)

df.head(20)
```

```python
# Rename nama_pedagang to nama_toko
df = df.rename(columns={'nama_pedagang': 'nama_toko'})
# Rename jenis_dagangan to kategori
df = df.rename(columns={'jenis_dagangan': 'kategori'})
df.head()
```

```python
# Display every unique value of kategori
df['kategori'].unique()
```

```python
categories = ['warung', 'buah dan sayur', 'sepatu', 'plastik', 'pakaian', 'elektronik', 'listrik', 'warung kopi', 'warung internet', 'penjahit', 'koran dan majalah', 'minuman', 'makanan', 'pulsa', 'kunci', 'warung tegal', 'jamu', 'aksesoris', 'kue', 'perabotan', 'obat', 'daging', 'emas', 'perhiasan', 'beras', 'jam tangan', 'pangkas rambut', 'galon']

# Replace every value of kategori with a random one from the categories list
import random
for i in range(len(df)):
    df.loc[i, 'kategori'] = random.choice(categories)
```

```python
# Randomize the order of the rows
df = df.sample(frac=1).reset_index(drop=True)
```

```python
# Reformat the nama_toko to be Capitalized for every word
df['nama_toko'] = df['nama_toko'].str.title()
```

```python
import numpy as np

# Add a new column called 'rerata_rating' and fill it with random values between 0 and 5, with one decimal place
df['rerata_rating'] = np.random.randint(0, 50, len(df)) / 10

# Add a new column called 'jumlah_rating' and fill it with random values between 0 and 50
df['jumlah_rating'] = np.random.randint(0, 50, len(df))
```

```python
# Remove every row that has a NaN value
df = df.dropna()

# Remove duplicate rows
df = df.drop_duplicates()
```

```python
# Add a new column called 'deskripsi' and fill it with null values
df['deskripsi'] = np.nan
```
